```python
#Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_usa = df[df['job_country'] == 'United States']
```

```python
df_usa.reset_index(inplace=True)
```

```python
df_indo = df[df['job_country'] == 'Indonesia']
df_indo.reset_index(inplace=True)
df_indo
```

```python
# List of ASEAN countries
asean_countries = [
    'Brunei', 'Cambodia', 'Indonesia', 'Laos', 'Malaysia', 
    'Myanmar', 'Philippines', 'Singapore', 'Thailand', 'Vietnam'
]

# Oceanic countries in the AANZFTA
oceanic_countries = [
    'Australia', 'New Zealand'
]

# Combined list for AANZFTA
aanzfta_countries = asean_countries + oceanic_countries

df_aanzfta = df[df['job_country'].isin(aanzfta_countries)]
df_aanzfta.reset_index(inplace=True)
df_aanzfta
```
